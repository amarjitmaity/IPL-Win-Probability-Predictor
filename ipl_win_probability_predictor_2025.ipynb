```python
# Keeping the rows where the team will be any team from "teams" list
match=match[match['team1'].isin(teams)]
match=match[match['team2'].isin(teams)]
match.head()
```

```python
## We don't want the matches effected by "Duckworth–Lewis–Stern (D/L) method" as the results of those matches are not actual
# Checking how many matches are effected by DLS and how many are not in the recent match dataframe
match['method'].unique()
```

```python
# Extracting the needed columns for getting our input variables from recent match dataframe
match=match[['id','city','winner','target_runs']]
match.head()
```

```python
# Merging the above extracted data with the delivery data
delivery_merged=match.merge(delivery,left_on='id',right_on='match_id')
delivery_merged.head()
```

```python
# Visualizing the no. of unique characters under the "inning" column of the delivery_merged dataframe above
delivery_merged['inning'].unique()
```

```python
# Extracting the data of 2nd innings only because we want the final outcomes during the 2nd innings
delivery_merged=delivery_merged[delivery_merged['inning']==2]
delivery_merged.head()
```

```python
# Calculating current scores after each balls of 2nd innings in the delivery_merged dataframe
delivery_merged['current_score']=delivery_merged.groupby('match_id')['total_runs'].cumsum()
delivery_merged.head()
```

```python
# Calculating the no. of runs left in the delivery_merged dataframe
delivery_merged['runs_left']=delivery_merged["target_runs"]-delivery_merged['current_score']
delivery_merged.head()
```

```python
# Calculating the no. of balls left in the delivery_merged dataframe
delivery_merged['balls_left']=120-(delivery_merged['over']*6+delivery_merged['ball'])
delivery_merged.head()
```

```python
# Calculating the no. of wickets left in the delivery_merged dataframe
delivery_merged['player_dismissed']=delivery_merged['player_dismissed'].fillna("0")
delivery_merged.head()
```

```python
delivery_merged['player_dismissed']=delivery_merged['player_dismissed'].apply(lambda x:x if x=="0" else "1")
delivery_merged['player_dismissed']=delivery_merged['player_dismissed'].astype('int')
delivery_merged.head()
```

```python
delivery_merged['wickets']=delivery_merged.groupby('match_id')['player_dismissed'].cumsum()
delivery_merged['wickets_left']=10-delivery_merged['wickets']
delivery_merged.head()
```

```python
# Calculating current run rates (crr=run/over) in the delivery_merged dataframe
delivery_merged['crr']=delivery_merged['current_score']/((120-delivery_merged['balls_left'])/6)
delivery_merged.head()
```

```python
# Calculating required run rates (rrr=run/over) in the delivery_merged dataframe
delivery_merged['rrr']=delivery_merged['runs_left']/((delivery_merged['balls_left'])/6)
delivery_merged.head()
```

```python
# Calculating results whether batting team has won or not in the delivery_merged dataframe
def result(row):
    return 1 if row['winner']==row['batting_team'] else 0
```

```python
delivery_merged['result']=delivery_merged.apply(result,axis=1)
delivery_merged.head()
```

```python
# Extracting the input features according to our need from the delivery_merged dataframe
final_df=delivery_merged[['batting_team','bowling_team','city','runs_left','balls_left',"wickets_left",'target_runs','crr','rrr','result']]
final_df.head()
```

```python
# Randomly shuffling the rows of the final_df dataframe to avoid biases while model training
final_df=final_df.sample(final_df.shape[0])
final_df.head()
```

We can see that the minimum of "rrr" column is "-inf" and the maximum of "rrr" column is 'inf' in the final_df dataframe

The "rrr" column contains "-inf" and "inf" values in those rows of the final_df dataframe, where there are some runs left but no balls left 

So, we'll remove those rows from the final_df dataframe, which have the no. of balls left as "0"

```python
# Removing all the rows from the final_df dataframe where "balls_left" is 0
final_df=final_df[final_df['balls_left']!=0]
```

```python
# Defining our independent and dependent variables in the final_df dataframe
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
```

```python
# Visualizing the top 5 rows of the independent variable matrix created above
x.head()
```

```python
# Visualizing the top 5 rows of the dependent variable created above
y.head()
```

```python
# Splitting our whole data into train data and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Transforming our categorical columns into numeric columns of the independent variable matrix
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ('trf1',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
trf
```

```python
# Creating the pipeline of ColumnTransformer and LogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe=Pipeline([
    ('steps1',trf),
    ('steps2',LogisticRegression(solver='liblinear'))
])
```

```python
# Training the model with x-values and y-values of the training dataset by logistic regression
print(pipe.fit(x_train,y_train))
```

```python
# Predicting the y-values from the x-values of the test dataset (logistic regression)
y_pred_test=pipe.predict(x_test)
```

```python
# Checking how accurately the model predicts the y-values from the x-xalues of the test dataset (logistic regression)
from sklearn.metrics import accuracy_score
acc_sc=accuracy_score(y_test,y_pred_test)
acc_sc
```

```python
# Checking the losing and winning probabilities for the chasing team at any certain point of time for the test data (logistic regression)
pipe.predict_proba(x_test)
```

Here if we want winning or losing percentage for the chasing team at any certain point of time, then logistic regression will give us the well-distributed probabilities in both outcomes

```python
## What will happen if we apply random forest instead of logistic regression
# Creating the pipeline of ColumnTransformer and RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
pipe1=Pipeline([
    ('steps1',trf),
    ('steps2',RandomForestClassifier())
])
```

```python
# Training the model with x-values and y-values of the training dataset by random forest classifier
print(pipe1.fit(x_train,y_train))
```

```python
# Predicting the y-values from the x-values of the test dataset (random forest classifier)
y_pred_test1=pipe1.predict(x_test)
```

```python
# Checking how accurately the model predicts the y-values from the x-xalues of the test dataset (random forest classifier)
from sklearn.metrics import accuracy_score
acc_sc1=accuracy_score(y_test,y_pred_test1)
acc_sc1
```

As this accuracy is very high, so we can predict whether the chasing team will win a match or not at a certain point of time with high confidence

But if we want winning or losing percentage for the chasing team at any certain point of time, then randomforest will give us the probabilities highly biased in one outcome in maximum cases

```python
# Checking the losing and winning probabilities for the chasing team at any certain point of time for the test data (random forest classifier)
pipe1.predict_proba(x_test)
```

So, we don't prefer random forest algorithm here
